```python
import cv2; import numpy as np; import matplotlib.pyplot as plt; import os; import matplotlib.cm as cm; import math; import mpl_toolkits.mplot3d.axes3d as p3; import pylab as p
```

```python
#display routines
def toPltImage(image):
    tmp = np.empty(image.shape,dtype=np.float64)
    if(len(image.shape) == 3):   
        tmp[:,:,0] = image[:,:,2]
        tmp[:,:,1] = image[:,:,1]
        tmp[:,:,2] = image[:,:,0]
    else:
        tmp[:] = image[:]
    return tmp / 255
def loadFromDir(directory,names):
    images = []
    for name in names:
        images.append(cv2.imread(directory + os.path.sep + name,cv2.CV_LOAD_IMAGE_UNCHANGED))
    return images

def imshow(im):
    if(len(im.shape) == 3):
        plt.imshow(toPltImage(im))
    else:
        plt.imshow(toPltImage(im),cmap = cm.Greys_r)

def overlay(background,overlay,opacity):
    if(len(overlay.shape) == 2):
        tmp = np.zeros((overlay.shape[0],overlay.shape[1],3),np.uint8)
        #make a yellow 3-channel version
        tmp[:,:,1] = overlay
        tmp[:,:,2] = overlay
        overlay = tmp
    if(len(background.shape) == 2):
        #convert to greyscale
        background = cv2.cvtColor(background,cv2.COLOR_GRAY2BGR)    
    #overlay = overlay.astype(np.float32) / 255
    #background = overlay.astype(np.float32) / 255
    return cv2.convertScaleAbs(cv2.addWeighted(overlay, opacity, background, 1 - opacity, 0))
```

```python
photoNames = os.listdir('images/photos')
photoNames.sort()
photos = loadFromDir('images/photos',photoNames)
sobels = loadFromDir('images/gradients',photoNames)
berkleys = loadFromDir('images/edges',photoNames)
```

```python
ix=24
photo = photos[ix]
sob = sobels[ix]
ber = berkleys[ix]
```

```python
gaussHalf = cv2.GaussianBlur(photo,(11,11),sigmaX = 0.5)
gauss5 = cv2.GaussianBlur(photo,(21,21),sigmaX =5)

diff = gaussHalf.astype(np.float32) - gauss5.astype(np.float32)

print diff.max()

diff += diff.min()

fig = plt.figure(figsize=(27.5,15))
imshow(diff)
```

```python
def intenseVar(im,kernelSize):
    kernel = np.ones((kernelSize,kernelSize),np.float64)
    N = kernelSize * kernelSize
    means = cv2.filter2D(im,-1,kernel) / N
    tmp = cv2.filter2D(np.power(im - means,4),-1,kernel) / N
    return np.max(tmp,axis=2)

src = photo.astype(np.float64) / 255
iv = intenseVar(src,3)

#disp = (iv / np.median(iv))
#fig = plt.figure(figsize=(27.5,15))
#plt.imshow(disp,cmap = cm.Greys_r)
fig = plt.figure(figsize=(27.5,15))
ax = p3.Axes3D(fig)
ax.plot_wireframe(np.arange(src.shape[0]),np.arange(src.shape[1]),iv)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('V')


    
```
