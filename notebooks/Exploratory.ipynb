```python
def imNameFromNumber(imageId):
    return "DSC_%04d.png" % imageId

def loadSamples(rowIx):
    featureVec = edges.ix[rowIx]
    imId = featureVec["image_id"]
    imName = imNameFromNumber(int(featureVec["image_id"]))
    edgeIndex = int(featureVec["edge_ix"])
    imageEdges = edges[edges["image_id"] == imId]
    startAt = imageEdges[imageEdges["edge_ix"]<edgeIndex]["edge_len"].sum()
    endAt = startAt + featureVec["edge_len"]
    samples = cv2.imread("images/samples/" + imName,cv2.CV_LOAD_IMAGE_UNCHANGED)
    samplingRange = samples.shape[0]/2
    sampleL = samples[0:samplingRange,startAt:endAt]
    sampleR = samples[samplingRange:samples.shape[0],startAt:endAt]
    return sampleL,sampleR
```

```python
def toPltImage(imageBGR):
    tmp = np.empty(imageBGR.shape,dtype=np.float64)
    tmp[:,:,0] = imageBGR[:,:,2]
    tmp[:,:,1] = imageBGR[:,:,1]
    tmp[:,:,2] = imageBGR[:,:,0]
    return tmp / 255
```

```python
def plotProfiles(edgeRowIndex):
    featureVec = edges.ix[edgeRowIndex]
    textureProfileL = lProfiles[edgeRowIndex << 1]
    textureProfileR = lProfiles[(edgeRowIndex << 1)+1]
    
    x = np.arange(len(textureProfileL))
    
    b1 = textureProfileL[:,0];b2 = textureProfileR[:,0]
    g1 = textureProfileL[:,1];g2 = textureProfileR[:,1]
    r1 = textureProfileL[:,2];r2 = textureProfileR[:,2]
    ave1 = textureProfileL.mean(axis=1); ave2 = textureProfileR.mean(axis=1)
    
    r = (1.0,0.0,0.0)
    g = (0.0,1.0,0.0)
    b = (0.0,0.0,1.0)
    black = (0,0,0)
    
    fig = plt.gcf()
    fig.set_size_inches(24,12)
    
    plt.subplot2grid((4,2), (0,0), colspan=1,rowspan=1); 
    plt.plot(x,r1,x,r2,c=r); plt.fill_between(x,r1,r2,alpha=0.1,color=r)
    plt.plot(x,g1,x,g2,c=g); plt.fill_between(x,g1,g2,alpha=0.1,color=g)
    plt.plot(x,b1,x,b2,c=b); plt.fill_between(x,b1,b2,alpha=0.1,color=b)
    
    plt.subplot2grid((4,2), (1,0), colspan=1,rowspan=1)
    plt.plot(x,ave1,c=black);plt.plot(x,ave2,c=(0.5,0.5,0.5))
    plt.fill_between(x,ave1,ave2,alpha=0.1,color=black)
    
    sampleL,sampleR = loadSamples(edgeRowIndex)
    plt.subplot2grid((4,2), (2,0), colspan=1)
    plt.imshow(toPltImage(sampleL))
    plt.subplot2grid((4,2), (3,0), colspan=1)
    plt.imshow(toPltImage(sampleR))
    
    imName = imNameFromNumber(int(featureVec["image_id"]))
    wi = core.WorkImage('images/photos/'+imName,'images/processed/'+imName,'images/belonging/'+imName,1023,imName,True)
    disp = np.copy(wi.image)
    wi.edges[int(featureVec["edge_ix"])].draw(disp,(0,255,255))
    plt.subplot2grid((4,2), (0,1), rowspan=4)
    plt.imshow(toPltImage(disp))
```
